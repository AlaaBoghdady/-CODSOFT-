```python
#Importing libraries
import pandas as pd
import numpy as np
import difflib
# لمقارنه النصوص و ايجاد التشابه
from sklearn.feature_extraction.text import TfidfVectorizer
# لتحويل الكلمات لتمثيل عددي
from sklearn.metrics.pairwise import cosine_similarity
#لقياس نسبه التشابه بين النصوص
```

```python
df = pd.read_csv("/content/movies.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# selecting Feature
imp_features = ['genres' , 'keywords' , 'tagline' , 'cast' , 'director']
print(imp_features)
```

```python
# replace null value with null string
for feature in imp_features:
  df[feature] = df[feature].fillna('')
```

```python
# combining the 5 features
combined_features = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
print(combined_features)
```

```python
# Convert text data to feature factors
vectorizer = TfidfVectorizer()
```

```python
feature_vactors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vactors)
```

Cosine Similarity

```python
# getting the similarity scores using Cosine Similarity

similarity = cosine_similarity(feature_vactors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

Getting the Movie Name from the user

```python
# Getting the Movie Name from the user
Movie_name = input('Enter your favourite movie name : ')
```

```python
#Creating a list of all movie names given in the dataset
list_of_titles = df['title'].tolist()
print(list_of_titles)
```

```python
# finding the close match for the movie name given by the user
close_match = difflib.get_close_matches(Movie_name ,list_of_titles)
print(close_match)
```

```python
match = close_match[0]
print(match)
```

```python
# Finding the index of the movie with title
index_movie = df[df.title == match]['index'].values[0]
print(index_movie)
```

```python
# getting a list of similar movies
similarity_score = list(enumerate(similarity[index_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score
sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print ('Movies suggested for you : \n')

i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_index = df[df.index == index ]['title'].values[0]
  if (i<30):
    print(i , '.' , title_index)
    i+=1
```

Movie Recommendation System

```python
movie_name = input ('Enter your favourite movie name :')

list_of_titles = df['title'].tolist()

close_match = difflib.get_close_matches(movie_name , list_of_titles)

match = close_match[0]

index_movie = df[df.title == match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_movie]))

sorted_similar_movies = sorted(similarity_score , key = lambda x:x[1] , reverse = True)

print('movies suggested for you : \n ')

i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_index = df[df.index == index]['title'].values[0]
  if (i<30):
    print(i , '.' , title_index)
    i+=1
```
